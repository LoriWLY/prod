```python
# Import libraries necessary for this project
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train_sample_pred= pd.read_csv("train_sample_pred.csv")
train_sample_pred
```

```python
plt.figure(figsize=(15,8))
ax = sns.kdeplot(train_sample_pred["y_pred"][train_sample_pred.recorded == 1], color="orange", shade=True)
sns.kdeplot(train_sample_pred["y_pred"][train_sample_pred.recorded == 0], color="teal", shade=True)
plt.legend(['reorder', 'non-reorder'])
plt.title('reorder or non-reorder by y_pred')
ax.set(xlabel='y_pred')
# plt.xlim(-0.01,1)
plt.show()
```

```python
train_data= pd.read_csv("train_data.csv")
```

```python
train_data.head()
```

```python
train_sample=train_data.sample(frac=0.05, random_state=None,axis=0)
train_sample.reset_index(drop=True, inplace=True)
train_sample
```
